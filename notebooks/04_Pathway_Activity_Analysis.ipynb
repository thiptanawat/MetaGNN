# Pathway Activity and Metabolic Feasibility Analysis

This notebook analyzes reaction activity scores, pathway-stratified distributions, and metabolic model feasibility. We examine how patient-specific metabolic networks differ in terms of active reactions and their pathway memberships.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path
import seaborn as sns

# Set matplotlib style for publication quality
mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['font.size'] = 10
mpl.rcParams['figure.dpi'] = 150
mpl.rcParams['savefig.dpi'] = 150

# Color scheme
colors = {
    'MetaGNN': '#2E75B6',
    'GIMME': '#A5A5A5',
    'iMAT': '#9DC3E6',
    'CORDA': '#BDD7EE',
    'tINIT': '#70AD47'
}

# Define paths
results_dir = Path('../results')
pathway_dir = results_dir / 'pathway_analysis'
fba_dir = results_dir / 'fba_feasibility'

print(f"Loading pathway and FBA data")
```

## Load Pathway Analysis Data

Load reaction activity scores and pathway-stratified distributions.

```python
# Load pathway analysis files
reaction_activity_dist = pd.read_csv(pathway_dir / 'reaction_activity_distribution.csv')
pathway_activity = pd.read_csv(pathway_dir / 'pathway_activity_per_patient.csv')

print("Reaction Activity Distribution:")
print(reaction_activity_dist.head(10))
print(f"Shape: {reaction_activity_dist.shape}")
print(f"Columns: {reaction_activity_dist.columns.tolist()}")

print("\nPathway Activity per Patient:")
print(pathway_activity.head())
print(f"Shape: {pathway_activity.shape}")
print(f"Columns: {pathway_activity.columns.tolist()}")
```

## Reaction Activity Score Distribution

Bimodal distribution showing active and inactive reactions.

```python
# Extract activity scores
if 'activity_score' in reaction_activity_dist.columns:
    activity_scores = reaction_activity_dist['activity_score']
elif 'score' in reaction_activity_dist.columns:
    activity_scores = reaction_activity_dist['score']
else:
    # Try to find numeric column
    numeric_cols = reaction_activity_dist.select_dtypes(include=[np.number]).columns
    activity_scores = reaction_activity_dist[numeric_cols[0]]

print(f"Activity Score Statistics:")
print(f"  Mean: {activity_scores.mean():.4f}")
print(f"  Std Dev: {activity_scores.std():.4f}")
print(f"  Min: {activity_scores.min():.4f}")
print(f"  Max: {activity_scores.max():.4f}")
print(f"  Median: {activity_scores.median():.4f}")
```

```python
# Plot activity score histogram
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(activity_scores, bins=40, color=colors['MetaGNN'], alpha=0.7, edgecolor='black')
ax.axvline(activity_scores.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {activity_scores.mean():.3f}')
ax.axvline(activity_scores.median(), color='green', linestyle='--', linewidth=2, label=f'Median: {activity_scores.median():.3f}')

# Mark bimodal peaks (approximately at 0 and 1 if binary-ish distribution)
ax.axvline(0.5, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label='Activity Threshold')

ax.set_xlabel('Reaction Activity Score', fontsize=11)
ax.set_ylabel('Frequency', fontsize=11)
ax.set_title('Reaction Activity Score Distribution', fontsize=12, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3, axis='y', linestyle='--')

plt.tight_layout()
plt.savefig('fig_activity_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Pathway-Stratified Activity Analysis

Compare reaction activity distributions across different metabolic pathways.

```python
# Prepare data for pathway stratification
# Assume pathway_activity has pathway column and activity scores
print("Pathway Activity Data Summary:")
if 'pathway' in pathway_activity.columns:
    print(f"Number of unique pathways: {pathway_activity['pathway'].nunique()}")
    print(f"Pathways: {pathway_activity['pathway'].unique()[:10].tolist()}")
    
    # Calculate statistics per pathway
    pathway_stats = pathway_activity.groupby('pathway').agg({
        'activity_score': ['mean', 'std', 'count']
    }).round(4)
    
    print("\nPathway Activity Statistics:")
    print(pathway_stats.head(10))
else:
    print("Pathway column not found. Available columns:")
    print(pathway_activity.columns.tolist())
```

```python
# Create pathway-stratified box plot
if 'pathway' in pathway_activity.columns and 'activity_score' in pathway_activity.columns:
    # Select top pathways by count for clarity
    top_pathways = pathway_activity['pathway'].value_counts().head(8).index
    pathway_data_filtered = pathway_activity[pathway_activity['pathway'].isin(top_pathways)]
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Create box plot
    bp = ax.boxplot([pathway_data_filtered[pathway_data_filtered['pathway'] == pathway]['activity_score'].values 
                      for pathway in top_pathways],
                     labels=top_pathways,
                     patch_artist=True,
                     notch=True)
    
    # Color the boxes
    for patch in bp['boxes']:
        patch.set_facecolor(colors['MetaGNN'])
        patch.set_alpha(0.7)
    
    ax.set_xlabel('Metabolic Pathway', fontsize=11)
    ax.set_ylabel('Activity Score', fontsize=11)
    ax.set_title('Reaction Activity Scores by Metabolic Pathway', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y', linestyle='--')
    ax.set_xticklabels(top_pathways, rotation=45, ha='right', fontsize=9)
    
    plt.tight_layout()
    plt.savefig('fig_pathway_activity.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## FBA Feasibility Analysis

Evaluate flux balance analysis results on patient-specific metabolic networks.

```python
# Load FBA summary
fba_summary = pd.read_csv(fba_dir / 'per_patient_fba_summary.csv')

print("Per-patient FBA Summary:")
print(fba_summary.head(10))
print(f"\nShape: {fba_summary.shape}")
print(f"Columns: {fba_summary.columns.tolist()}")

# Print statistics
numeric_cols = fba_summary.select_dtypes(include=[np.number]).columns
print("\nFBA Statistics:")
print(fba_summary[numeric_cols].describe().round(4))
```

```python
# Visualize FBA metrics
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Get numeric columns for visualization
numeric_cols = fba_summary.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) >= 2:
    # Reactions Retained
    ax = axes[0, 0]
    if 'reactions_retained' in numeric_cols:
        reactions_retained = fba_summary['reactions_retained']
    else:
        reactions_retained = fba_summary[numeric_cols[0]]
    
    ax.hist(reactions_retained, bins=20, color=colors['MetaGNN'], alpha=0.7, edgecolor='black')
    ax.axvline(reactions_retained.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {reactions_retained.mean():.1f}')
    ax.set_xlabel('Number of Reactions', fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)
    ax.set_title('Reactions Retained per Patient', fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3, axis='y', linestyle='--')
    
    # Biomass Flux
    ax = axes[0, 1]
    if 'biomass_flux' in numeric_cols:
        biomass_flux = fba_summary['biomass_flux']
    else:
        biomass_flux = fba_summary[numeric_cols[1]] if len(numeric_cols) > 1 else fba_summary[numeric_cols[0]]
    
    ax.hist(biomass_flux, bins=20, color='#E74C3C', alpha=0.7, edgecolor='black')
    ax.axvline(biomass_flux.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {biomass_flux.mean():.3f}')
    ax.set_xlabel('Biomass Flux', fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)
    ax.set_title('Biomass Flux per Patient Model', fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3, axis='y', linestyle='--')
    
    # Correlation between reactions retained and biomass flux
    ax = axes[1, 0]
    ax.scatter(reactions_retained, biomass_flux, color=colors['MetaGNN'], alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
    
    # Add trend line
    z = np.polyfit(reactions_retained, biomass_flux, 1)
    p = np.poly1d(z)
    ax.plot(reactions_retained, p(reactions_retained), "r--", linewidth=2, label='Trend')
    
    ax.set_xlabel('Reactions Retained', fontsize=10)
    ax.set_ylabel('Biomass Flux', fontsize=10)
    ax.set_title('Correlation: Model Size vs. Biomass Production', fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Scatter plot of all patients
    ax = axes[1, 1]
    patient_idx = np.arange(len(fba_summary))
    ax.scatter(patient_idx, reactions_retained, alpha=0.6, s=50, color=colors['MetaGNN'], label='Reactions Retained')
    ax.set_xlabel('Patient Index', fontsize=10)
    ax.set_ylabel('Reactions Retained', fontsize=10)
    ax.set_title('Model Complexity Across Patients', fontsize=11, fontweight='bold')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.legend(fontsize=9)
    
    plt.tight_layout()
    plt.savefig('fig_fba_analysis.png', dpi=150, bbox_inches='tight')
    plt.show()
```

## Metabolic Task Completion

Analysis of whether patient-specific networks can perform essential metabolic tasks.

```python
# Load task completion detail
task_completion_detail = pd.read_csv(fba_dir / 'metabolic_task_completion_detail.csv')

print("Metabolic Task Completion Details:")
print(task_completion_detail.head(15))
print(f"\nShape: {task_completion_detail.shape}")
print(f"Columns: {task_completion_detail.columns.tolist()}")
```

```python
# Create task completion heatmap or summary
# Try to create a pivot table if we have task and patient identifiers

# Check structure
if 'task' in task_completion_detail.columns or 'task_id' in task_completion_detail.columns:
    task_col = 'task' if 'task' in task_completion_detail.columns else 'task_id'
    
    # Get completion status column
    completion_cols = [col for col in task_completion_detail.columns if 'completion' in col.lower() or 'feasible' in col.lower()]
    if completion_cols:
        completion_col = completion_cols[0]
    else:
        completion_col = task_completion_detail.select_dtypes(include=[bool, np.number]).columns[0]
    
    # Count task completion
    task_summary = task_completion_detail.groupby(task_col)[completion_col].agg(['sum', 'count'])
    task_summary['completion_rate'] = task_summary['sum'] / task_summary['count']
    
    print("\nTask Completion Summary:")
    print(task_summary.sort_values('completion_rate', ascending=False))
    
    # Visualize
    fig, ax = plt.subplots(figsize=(12, 6))
    
    tasks = task_summary.index
    completion_rates = task_summary['completion_rate']
    
    bars = ax.barh(tasks, completion_rates, color=colors['MetaGNN'], alpha=0.7, edgecolor='black')
    ax.set_xlabel('Completion Rate', fontsize=11)
    ax.set_ylabel('Metabolic Task', fontsize=11)
    ax.set_title('Metabolic Task Completion Rates Across Patients', fontsize=12, fontweight='bold')
    ax.set_xlim([0, 1.05])
    ax.grid(True, alpha=0.3, axis='x', linestyle='--')
    
    # Add value labels
    for bar, rate in zip(bars, completion_rates):
        ax.text(rate + 0.02, bar.get_y() + bar.get_height()/2, f'{rate:.2%}',
                va='center', fontsize=9)
    
    plt.tight_layout()
    plt.savefig('fig_task_completion.png', dpi=150, bbox_inches='tight')
    plt.show()
else:
    # Create a summary of whatever structure we have
    print("Task completion data structure:")
    print(task_completion_detail.describe())
```
