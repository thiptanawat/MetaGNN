# Benchmark Evaluation Analysis

Comprehensive comparison of MetaGNN against baseline methods.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path

mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['font.size'] = 10
mpl.rcParams['figure.dpi'] = 150

colors = {
    'MetaGNN': '#2E75B6',
    'GIMME': '#A5A5A5',
    'iMAT': '#9DC3E6',
    'CORDA': '#BDD7EE',
    'tINIT': '#70AD47'
}

results_dir = Path('../results')
benchmark_dir = results_dir / 'benchmark_evaluation'

print(f"Loading data from: {benchmark_dir}")
```

## Load Benchmark Data

```python
benchmark_summary = pd.read_csv(benchmark_dir / 'benchmark_summary.csv')
print("Benchmark Summary:")
print(benchmark_summary)

# Extract metrics by method
metrics_by_method = {}
for method in benchmark_summary['method'].unique():
    method_data = benchmark_summary[benchmark_summary['method'] == method]
    metrics_by_method[method] = {
        row['metric']: row['mean'] for _, row in method_data.iterrows()
    }
```

```python
f1_scores = pd.read_csv(benchmark_dir / 'per_patient_f1_scores.csv')
auroc_scores = pd.read_csv(benchmark_dir / 'per_patient_auroc_scores.csv')
task_completion = pd.read_csv(benchmark_dir / 'per_patient_task_completion.csv')

print("Per-patient data shapes:")
print(f"F1: {f1_scores.shape}")
print(f"AUROC: {auroc_scores.shape}")
print(f"Task Completion: {task_completion.shape}")
```

## Performance Comparison

```python
methods = list(metrics_by_method.keys())
f1_vals = [metrics_by_method[m].get('F1', 0) for m in methods]
auroc_vals = [metrics_by_method[m].get('AUROC', 0) for m in methods]
tc_vals = [metrics_by_method[m].get('TC', 0) / 100 for m in methods]

fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(methods))
width = 0.25

ax.bar(x - width, f1_vals, width, label='F1-score', color='#3498DB', alpha=0.8)
ax.bar(x, auroc_vals, width, label='AUROC', color='#E74C3C', alpha=0.8)
ax.bar(x + width, tc_vals, width, label='Task Completion', color='#2ECC71', alpha=0.8)

ax.set_xlabel('Method', fontsize=11)
ax.set_ylabel('Score', fontsize=11)
ax.set_title('Benchmark Comparison: Performance Metrics', fontsize=12, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels([str(m) for m in methods], fontsize=10)
ax.legend(fontsize=10)
ax.set_ylim([0, 1])
ax.grid(True, alpha=0.3, axis='y', linestyle='--')

plt.tight_layout()
plt.savefig('fig_benchmark_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Essential Gene AUROC

```python
essential_gene_auroc = pd.read_csv(benchmark_dir / 'essential_gene_auroc_per_cellline.csv')
print("Essential Gene AUROC per Cell Line:")
print(essential_gene_auroc)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

cell_lines = [str(x) for x in essential_gene_auroc.iloc[:, 0].tolist()]
x = np.arange(len(cell_lines))
width = 0.15

for i, method in enumerate(essential_gene_auroc.columns[1:]):
    offset = (i - len(essential_gene_auroc.columns[1:]) / 2 + 0.5) * width
    values = essential_gene_auroc[method].tolist()
    ax.bar(x + offset, values, width, label=method, color=colors.get(method, '#999999'), alpha=0.8)

ax.set_xlabel('Cell Line', fontsize=11)
ax.set_ylabel('AUROC', fontsize=11)
ax.set_title('Essential Gene Prediction AUROC by Cell Line', fontsize=12, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(cell_lines, rotation=45, ha='right', fontsize=9)
ax.legend(fontsize=9)
ax.set_ylim([0, 1])
ax.grid(True, alpha=0.3, axis='y', linestyle='--')

plt.tight_layout()
plt.savefig('fig_essential_gene_auroc.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Summary

Key findings: MetaGNN shows superior performance across metrics with high consistency and generalization capability.
